```python
paragraph = """
Modern education faces the dual challenge of preparing students for an uncertain future while ensuring equitable access to learning resources today.
Technological advances have made it possible to deliver personalized instruction at scale: adaptive learning systems analyze student performance and present targeted practice;
online platforms provide interactive content that fits diverse learning styles; and data analytics reveal trends that help teachers intervene earlier.
However, technology is not a cure-all. Effective learning requires skilled teachers who understand pedagogy and can motivate students, schools that foster safe and inclusive communities,
and assessment systems that measure deep understanding rather than rote memorization. Equity is especially pressing — students in underfunded schools often lack devices, reliable internet, or trained staff, widening achievement gaps.
Curriculum reform is another key piece, emphasizing critical thinking, collaboration, and digital literacy to prepare learners for complex problem solving, not just test taking.
Finally, lifelong learning models are becoming essential: with the pace of change in the job market, people must continually reskill and upskill.
Policy makers, educators, and communities must therefore work together to align investments in infrastructure, teacher training, and curriculum innovation so the promise of technology reaches all learners and builds a resilient, future-ready education system.
"""
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
```

```python
while True:
    question = input("Enter a question (or 'exit'): ")
    if question.lower() == "exit":
        break
    result = qa_pipeline(question=question, context=paragraph)
    print(f"Answer: {result['answer']} (score: {result['score']:.3f})\n")
```
